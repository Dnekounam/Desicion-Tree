```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/danialnekounam/Desktop/car.csv')
data
```

```python
dummies=pd.get_dummies(data,columns=['vhigh','vhigh.1','2','2.1','small','low'])
dummies
```

```python
 print ("Dataset: ",dummies.head())
```

```python
X = dummies.values[:, 1:21]
Y = dummies.values[:, 0]
df = pd.DataFrame(dummies)
dummies
#col_name='low_med'
#index_no = df.columns.get_loc(col_name)
#print("Index of {} column in given dataframe is : {}".format(col_name, index_no))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( 
X, Y, test_size = 0.3, random_state = 100)
X_train 
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_gini = DecisionTreeClassifier(criterion = "gini",
            random_state = 100,max_depth=10, min_samples_leaf=21)
clf_gini.fit(X_train, y_train)
clf_gini.score(X_test,y_test)
```

```python
clf_entropy = DecisionTreeClassifier(
            criterion = "entropy", random_state = 100,
            max_depth = 10, min_samples_leaf = 21)
clf_entropy.fit(X_train, y_train)
clf_entropy
```

```python
y_pred = clf_entropy.predict(X_test)
print("Predicted values:")
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
print("Confusion Matrix: ",
        confusion_matrix(y_test, y_pred))
      
print ("Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
      
print("Report : ",
    classification_report(y_test, y_pred))
  
```
